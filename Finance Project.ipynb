**Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()

for t2 in tickers:
    returns[t2+' return'] = bank_stocks[t2]['Close'].pct_change()
    returns['Date'] = bank_stocks[t2]['Date']
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
# Creating a Close graph for BAC in 2008
BAC_Close_2008 = bank_stocks2.loc['2008']['Close']['BAC']

# Creating a 30-day moving average for BAC in 2008
AVG_2008 = bank_stocks2.loc['2008']['Close'].rolling(window=30).mean()
BAC_AVG_2008 = AVG_2008['BAC']

# Convert the datetime index to strings to avoid the deprecation warning
BAC_Close_2008.index = BAC_Close_2008.index.astype(str)
BAC_AVG_2008.index = BAC_AVG_2008.index.astype(str)

# Combine both the closing prices and the moving average into one DataFrame
combined_df = pd.DataFrame({
    'BAC Close (2008)': BAC_Close_2008,
    'BAC 30-Day Moving Avg (2008)': BAC_AVG_2008
})

# Plot both on the same interactive graph using iplot
combined_df.iplot(title='BAC Closing Prices and 30-Day Moving Average (2008)',
                  xTitle='Date',
                  yTitle='Price',
                  colors=['darkblue', 'green'])
```

**Create a heatmap of the correlation between the stocks Close Price.**

```python
## Create a correaltin matrix variable for all tickers
corr_matrix = bank_stocks2['Close'].corr()

plt.figure(figsize=(12, 8)) ## Increase the size of the graph

## Create a heatmap corr. for all tickers
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation of Bank Stock Closing Prices (2006-2016)', fontsize=16, fontweight='bold',y=1.05,color='darkblue')

# Add x_label
plt.xlabel('Ticker', fontweight='bold')

# Add y_label
plt.ylabel('Ticker', fontweight='bold')

plt.show()
```

**Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
plt.figure(figsize=(12, 8)) ## Increase the size of the graph

## Create a heatmap corr. for all tickers
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.show()
```
